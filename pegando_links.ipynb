```python
navegador = webdriver.Chrome()

lista_links = list()

url_selenium = "https://www.glassdoor.com.br/Job/Home/recentActivity.htm"

navegador.get(url_selenium)

navegador.find_element_by_xpath('//*[@id="userEmail"]').send_keys(e_mail)
navegador.find_element_by_xpath('//*[@id="userPassword"]').send_keys(senha)
navegador.find_element_by_xpath('//*[@id="InlineLoginModule"]/div/div/div/div[1]/div[3]/form/div[3]/div[1]/button').click()
time.sleep(10)

url_selenium = "https://www.glassdoor.com.br/Sal%C3%A1rios/fortaleza-sal%C3%A1rio-SRCH_IL.0,9_IM3050.htm?clickSource=searchBtn"
navegador.get(url_selenium)

navegador.find_element_by_xpath('//*[@id="sc.location"]').click()
pyautogui.press("end")
for delete in range (1, 100):
    pyautogui.press("backspace")

    
df_capitais = pd.read_excel("links_capitais.xlsx")

for capital in range(0, len(df_capitais["Capitais"]) + 1):
    local = f'{df_capitais["Capitais"][capital]}, {df_capitais["Estados"][capital]}, Brasil'
    
    print(">>>>> ================ <<<<<")
    print(f"rodada: {capital}")
    print(f"local: {local}")
    
    navegador.find_element_by_xpath('//*[@id="sc.location"]').click()
    navegador.find_element_by_xpath('//*[@id="sc.location"]').send_keys(local)
    navegador.find_element_by_xpath('//*[@id="sc.location"]').send_keys(Keys.ENTER)  
    time.sleep(10)
    navegador.find_element_by_xpath('//*[@id="nodeReplace"]/div/div/div[2]/div[4]/div[1]/div[3]/div/div[1]/button[2]').click()
    navegador.find_element_by_xpath('//*[@id="sc.location"]').click()
    
    print(navegador.current_url)
    lista_links.append(navegador.current_url)
    
    for delete in range (1, len(local) + 1):
        pyautogui.press("end")
        pyautogui.press("backspace")

navegador.quit()
#navegador.find_element_by_xpath('//*[@id="sc.location"]').clear()
#navegador.find_element_by_xpath('//*[@id="sc.location"]').send_keys()
```
